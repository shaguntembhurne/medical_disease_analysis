```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import sklearn 
```

```python
data = pd.read_csv('/Users/shaguntembhurne/med/dataset.csv')
des = pd.read_csv('/Users/shaguntembhurne/med/symptom_Description.csv')
pre = pd.read_csv('/Users/shaguntembhurne/med/symptom_precaution.csv')
sym = pd.read_csv('/Users/shaguntembhurne/med/Symptom-severity.csv')
print('dataset overview:')
print(data.head())
print('\ncolumns:',data.columns)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Combine all symptoms into a single list per row, ignoring NaN values
symptom_columns = data.columns[1:]  # Excluding 'Disease'
data['symptoms'] = data[symptom_columns].apply(lambda row: [s for s in row if isinstance(s, str) and s.strip()], axis=1)

# Apply MultiLabelBinarizer for encoding
mlb = MultiLabelBinarizer()
encoded_symptoms = pd.DataFrame(mlb.fit_transform(data['symptoms']), columns=mlb.classes_)

# Concatenate encoded symptoms with Disease column
final_data = pd.concat([data[['Disease']], encoded_symptoms], axis=1)

# Display the first few rows of the encoded dataset
final_data.head()
```

```python
from sklearn.model_selection import train_test_split
x = encoded_symptoms
y = final_data['Disease']
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,train_size=0.2,random_state=42)
```

using random classifier model 


```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
print('accuracy:',accuracy_score(y_test,y_pred))
```

```python
from sklearn.svm import SVC
# Step 1: Initialize the SVM model
svm_model = SVC(kernel='linear')  # You can also try 'rbf' or 'poly' kernel

# Step 2: Train the model
svm_model.fit(X_train, y_train)

# Step 3: Make predictions
y_pred_svm = svm_model.predict(X_test)

# Step 4: Evaluate the model
accuracy = accuracy_score(y_test, y_pred_svm)
print("SVM Accuracy:", accuracy)
```

```python
import numpy as np

# Function to predict disease based on symptoms
def predict_disease(symptoms, model, all_symptoms):
    # Create an array of zeros (same size as training features)
    input_data = np.zeros(len(all_symptoms))

    # Mark the given symptoms as 1 in the input array
    for symptom in symptoms:
        if symptom in all_symptoms:
            input_data[all_symptoms.index(symptom)] = 1

    # Reshape and predict
    prediction = model.predict([input_data])
    
    return prediction[0]  # Return predicted disease

# Example usage
user_symptoms = ["itching", "skin_rash"]  # Replace with user input
predicted_disease = predict_disease(user_symptoms, svm_model, list(x.columns))

print("Predicted Disease:", predicted_disease)
```
